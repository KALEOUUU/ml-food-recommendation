## Categorize dataset

```python
def categorize_complete_meal_type(row):
    name = row["name"].lower()
    
    # Drinks
    if any(x in name for x in ["juice", "coffee", "tea", "milk", "soda", "beer", "wine", "drink", "beverage", 'kopi', 'teh', 'sirup', 'sari kedelai', 'es', 'jus', 
                               'susu', 'cendol', 'daun teh', 'bir', 'limun', 'bandrek', 'kelapa', 'nectar', 'fruit', 'cola', 'coke', 'tonic', 'water', 'coconut', 'sprite', 'limeade', 'lemonade']):
        return "Drink"
    
    # Carbs
    elif any(x in name for x in ["rice", "corn", "potato", "bread", "pasta", "noodle", "cereal", "oat", "wheat", "barley", 'nasi', 
                                 'roti', 'kentang', 'jagung', 'beras', 'kacang', 'talas', 'sukun', 'ubi', 'umbi', 'bihun', 'makaroni', 'tapioka', 'singkong', 'emping', 'gaplek', 'tepung', 'ketan', 'sagu', 'mie', 'bakso', 'lontong']):
        return "Carbs"
    
    # Breakfast
    elif any(x in name for x in ["oat", "cereal", "bread", "toast", "pancake", "waffle", "muffin", "yogurt", "egg", 'ketan', 'bubur', 'biskuit', 'kue', 'martabak', 'serabi', 'lupis', 'roti bakar', 'peuyeum']):
        return "Breakfast"
    
    # Lunch
    elif any(x in name for x in ["rice", "pasta", "noodle", "beef", "chicken", "fish", "soup", "vegetable", "lentil", "stew", 'bebek', 'ayam', 'ikan', 'sop', 'mi', 'mie', 'sayur', 'rendang', 'soto', 'telur', 'tahu', 'tempe', 'gulai', 
                                 'sayur asem', 'bakmi', 'pecel', 'opor', 'lodeh', 'rawon', 'tongseng', 'gudeg', "salad", "soup", "grilled", "steak", "seafood", "fish", "lentil", 'sup', 'ikan', 'steak', 'pecel', 'gado-gado', 
                                 'capcay', 'asinan', 'karedok', 'urap', 'daging', 'sapi', 'sate', 'abon', 'empal', 'sate', 'belut', 'kangkung', 'coto', 'koro', 'teri', 'sphagetty', 'semur']):
        return "Lunch/Dinner"

    # Snack
    elif any(x in name for x in ["cookie", "candy", "chocolate", "ice cream", "cracker", "chip", "bar", "snack", "popcorn", 'keripik', 'kerupuk', 'kue', 'pisang', 'jambu', 'jeruk', 'karoket', 'kelepon', 'kembang', 'ketela', 'onde-onde', 
                                 'bika', 'dodol', 'wingko', 'es', 'lemper', 'arem-arem', 'kue cubit', 'nastar', 'kastengel', 'lapis legit', 'mochi', 'getuk', 'cucur', 'biji salak', 'misro', 'combro', 'pudding', 'buah', 'mangga', 'markisa', 
                                 'ongol', 'pepaya', 'rambutan', 'siomay', 'tahu', 'tempe', 'granola', 'alpukat', 'pempek', 'oncom', 'lopis', 'manggis', 'nanas', 'nangka', 'olah-olah', 'papeda', 'pastel', 'ongol-ongol', 'sawo', 'srikaya', 'semangka', 'peanuts', 'seed', 'roasted', 'kelapa', 'wortel']):
        return "Snack"

    return "ingrideint"

df_general["Meal Type"] = df_general.apply(categorize_complete_meal_type, axis=1)
df_general["Meal Type"].value_counts()
```
